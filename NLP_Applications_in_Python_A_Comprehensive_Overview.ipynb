## 1. Using spaCy for Named Entity Recognition (NER)

````python
# Install spaCy: Uncomment the line below and run it first
!pip install spacy

# Run the following commands in a separate cell, or after installation is complete
import spacy

# After installation, you must also download the specific language model:
!python -m spacy download en_core_web_sm

# Load pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "Barack Obama was born on August 4, 1961, in Honolulu, Hawaii."

# Process text
doc = nlp(text)

# Extract entities
print("Entities and their labels: - NLP Applications 1.ipynb:20")
for ent in doc.ents:
    print(f"{ent.text} > {ent.label_} - NLP Applications 1.ipynb:22")
````

## 2. Using NLTK for Text Tokenization

````python
# Install NLTK: !pip install nltk
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

# Input text
text = "Natural Language Processing is exciting! Let's learn it together."

# Tokenize text into words
tokens = nltk.word_tokenize(text)
print("Word Tokens:", tokens)

# Tokenize text into sentences
sentences = nltk.sent_tokenize(text)
print("Sentence Tokens:", sentences)
````

## 3. Using Hugging Face Transformers for Sentiment Analysis

````python
# Install Hugging Face Transformers: !pip install transformers
from transformers import pipeline

# Load pre-trained sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Input text
text = ["I love this product!", "I hate waiting in line."]

# Perform sentiment analysis
results = sentiment_pipeline(text)
for i, result in enumerate(results):
    print(f"Text: {text[i]}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}")
````

## 4. Using spaCy for Dependency Parsing

````python
# Dependency Parsing with spaCy
doc = nlp("The quick brown fox jumps over the lazy dog.")
print("Dependency Parsing:")
for token in doc:
    print(f"Token: {token.text}, Dependency: {token.dep_}, Head: {token.head.text}")
````

## 5. Using NLTK for Stemming and Lemmatization

````python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Example words
words = ["running", "runner", "ran", "easily", "flying"]

# Apply stemming and lemmatization
print("Stemming Results:")
for word in words:
    print(f"{word} -> {stemmer.stem(word)}")

print("\nLemmatization Results:")
for word in words:
    print(f"{word} -> {lemmatizer.lemmatize(word, pos=wordnet.VERB)}")
````

## 6. Using Hugging Face for Text Generation (GPT-2)

````python
from transformers import pipeline

# Load pre-trained text generation pipeline
generator = pipeline("text-generation", model="gpt2")

# Generate text
prompt = "Once upon a time in a distant land"
generated_text = generator(prompt, max_length=50, num_return_sequences=1)
print("Generated Text:")
print(generated_text[0]['generated_text'])
````

## 7. Using spaCy for Part-of-Speech Tagging

````python
# Part-of-Speech Tagging with spaCy
doc = nlp("Apple is looking at buying a U.K. startup.")
print("Part-of-Speech Tags:")
for token in doc:
    print(f"Token: {token.text}, POS: {token.pos_}, Tag: {token.tag_}")
````

## 9. Using Hugging Face for Text Summarization

````python
from transformers import pipeline

# Load the summarization pipeline with a specific model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input text
text = """The 3MTT DeepTech Ready Program, spearheaded by Data Science Nigeria with support from Google,
is an innovative initiative designed to empower individuals with cutting-edge skills in technology. This
program focuses on fostering expertise in areas such as artificial intelligence, machine learning, and data
analytics, preparing participants to excel in the rapidly evolving tech landscape. Through hands-on training,
mentorship from industry experts, and real-world project implementation, the program aims to bridge the gap between
theoretical knowledge and practical application.
By equipping learners with these advanced skills, the 3MTT DeepTech Ready Program is contributing to building a robust
and competitive tech talent pool in Nigeria."""

# Generate summary
summary = summarizer(text, max_length=40, min_length=20, do_sample=False)
print("Summary:")
print(summary[0]['summary_text'])
````

## 11. Using NLTK for Word Frequency Analysis

````python
from nltk.probability import FreqDist

# Input text
text = "Natural Language Processing enables machines to understand human language."

# Tokenize and analyze frequency
tokens = nltk.word_tokenize(text.lower())
freq_dist = FreqDist(tokens)

# Display frequencies
print("Word Frequencies:")
for word, frequency in freq_dist.most_common():
    print(f"{word}: {frequency}")
````

## 14. Word Frequency Visualization

````python
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
import nltk

# Input text (reusing from previous example)
text_for_freq = "Natural Language Processing enables machines to understand human language."

# Tokenize and analyze frequency
tokens_for_freq = nltk.word_tokenize(text_for_freq.lower())
freq_dist_for_freq = FreqDist(tokens_for_freq)

# Get the most common words and their frequencies
most_common_words = freq_dist_for_freq.most_common(5) # Get top 5 words
words = [word for word, frequency in most_common_words]
frequencies = [frequency for word, frequency in most_common_words]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 5 Word Frequencies')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

## 13. Text Classification with Scikit-learn and NLTK

````python
# Install necessary libraries: !pip install scikit-learn

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
import re

# Sample dataset for text classification
data = [
    ("I love this movie, it's fantastic!", "positive"),
    ("What a terrible film, I hated it.", "negative"),
    ("This product is amazing, highly recommend.", "positive"),
    ("The service was slow and inefficient.", "negative"),
    ("Great book, very insightful and engaging.", "positive"),
    ("Worst experience ever, complete waste of money.", "negative"),
    ("Enjoyed every moment, truly captivating.", "positive"),
    ("Disappointing performance, lacked originality.", "negative")
]

texts = [item[0] for item in data]
labels = [item[1] for item in data]

# Preprocessing function (reusing from topic modeling example)
def preprocess_text(text):
    text = text.lower() # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text) # Remove punctuation and numbers
    tokens = nltk.word_tokenize(text) # Tokenize
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words] # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens] # Lemmatize
    return " ".join(tokens)

preprocessed_texts = [preprocess_text(text) for text in texts]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(preprocessed_texts, labels, test_size=0.3, random_state=42)

# Create a pipeline with TF-IDF vectorizer and a Naive Bayes classifier
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("\nText Classification Results:")
print(metrics.classification_report(y_test, y_pred))

# Example of prediction on new text
new_texts = [
    "This is a wonderful day!",
    "I am so frustrated with this situation."
]
preprocessed_new_texts = [preprocess_text(text) for text in new_texts]
predictions = model.predict(preprocessed_new_texts)

print("\nPredictions for new texts:")
for text, pred in zip(new_texts, predictions):
    print(f"Text: '{text}' -> Predicted: {pred}")
````

## 12. Topic Modeling with Scikit-learn and NLTK

````python
# Install necessary libraries: !pip install scikit-learn

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
import re

# Sample documents
documents = [
    "Machine learning is a field of artificial intelligence that uses statistical techniques.",
    "Natural Language Processing is a field of artificial intelligence that focuses on the interaction between computers and human language.",
    "Deep learning is a subset of machine learning based on artificial neural networks.",
    "Text analysis involves extracting meaningful insights from textual data.",
    "Data science combines various fields to understand and analyze data."
]

# Preprocessing function
def preprocess_text(text):
    text = text.lower() # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text) # Remove punctuation and numbers
    tokens = nltk.word_tokenize(text) # Tokenize
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words] # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens] # Lemmatize
    return " ".join(tokens)

preprocessed_documents = [preprocess_text(doc) for doc in documents]

# Create a CountVectorizer to convert text to a matrix of token counts
vectorizer = CountVectorizer()
data_vectorized = vectorizer.fit_transform(preprocessed_documents)

# Apply Latent Dirichlet Allocation (LDA) for topic modeling
num_topics = 2  # You can adjust this number
lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_model.fit(data_vectorized)

# Display the top words for each topic
print("\nTopics found by LDA:")
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda_model.components_): # Corrected from components__ to components_
    print(f"Topic #{topic_idx + 1}:")
    print(" ".join([feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]))
````
